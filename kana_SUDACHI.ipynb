# モデル定義

```
c_path = './DATA/Kessen.txt'

c_df = copy.deepcopy(df)

with open(c_path, 'a') as f:
    for index, row in c_df.iterrows():
        parsed = jaconv.kata2hira(row['回答語形'])
        if parsed == "" :
            parsed = row['回答語形']
        row['A'] = parsed
        
        #print(parsed,file=f)
c_df = c_df.reindex(columns=['項目名', 'A', '市区町村', '回答語形'])
print(c_df.head())

c_df.to_csv('./DATA/Tugaru2.csv', index=False)
```
